```python
import seaborn as sns
#sns.boxplot(x=df_new1['dep_delay'])
from matplotlib import pyplot as plt
plt.scatter(df_new1['dep_delay'],df_new1["month"])
```

```python
#signed number of std by which an observation is above mean value
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df_new1["dep_delay"]))
print(np.where(z>3))
```

```python
#IQR - difference between 75th and 25th percentiles
Q1 = df_new1.quantile(0.25)
Q3 = df_new1.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df_new2 = (df_new1 < (Q1 - 1.5 * IQR)) |(df_new1 > (Q3 + 1.5 * IQR))
df_new2[100:200]
```

```python
#Removing outliers using IQR score
df_new1 = df_new1[~((df_new1 < (Q1 - 1.5 * IQR)) |(df_new1 > (Q3 + 1.5 * IQR))).any(axis=1)]
df_new1
```

```python
low_information_cols=[]
num_rows = df_new1.shape[0]
for col in df_new1.columns:
    cnts = df_new1[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    
    if top_pct > 0.95:
        low_information_cols.append(col)
        print('{0}: {1:.5f}%'.format(col, top_pct*100))
        print(cnts)
        print()
```
