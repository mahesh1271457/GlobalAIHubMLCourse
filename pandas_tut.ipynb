Pandas is an opensource python library used for data analysis and data manipulation

Steps for analysing - load, prepare, manipulate, model ad analyze

Three ways to store the data: 
    1. Panel --> 3d structure data.
    2. Series --> 1d structure data.
    3. Dataframe --> 2d structure to store the data.



```python
#Python Pandas series
import pandas as pd
import numpy as np

#Empty data series
data = pd.Series()

#convert an ndarray into a data series
data_array = np.array(['a','b','c','d'])
data = pd.Series(data_array, index=[1,2,3,4])
print("Ndarray into data series:\n",data)

#convert a dictionary into a data series
data_dict = {'a':1,'b':2,'c':3,'d':4}
data = pd.Series(data_dict, index=['b','d','a','e','c'])
print("\n Dictionary into data series:\n",data)

#To access an element in data series
print("\nElement in first data series is : ",data[0])

#Retrieving data with label
print("\nElement in first data series is : ",data['a'])
```

```python
#Read data from pandas into a dataframe through a file

import pandas as pd

#Reading data from a file into a dataframe 
#df = pd.read_csv("filename")

#list to a dataframe
data = [1,2,3,4,5]
df = pd.DataFrame(data)
#print("list to dataframe:")
#print(df)

#list of lists to a dataframe
data = [['Alex',10],['Bob',12],['Clarke',13]]
df = pd.DataFrame(data,columns=['Name','Age'],dtype='float')
#print("\nlist of lists to dataframe:")
#print(df)

#dictionary of lists and series to a dataframe
data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
df = pd.DataFrame(data)
#print("\ndictionary of lists and series to a dataframe:")
#print(df)

#list of dictionaries into a dataframe
data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
df = pd.DataFrame(data)
#print("\nlists of dictionaries to a dataframe:")
#print(df)

#dataframe addition
df['d'] = df['a'] + df['b']
#print(df)

#dataframe deletion
del(df['d'])
print(df)

#selecting a value with an id
df.loc[0]

#selecting a value 
df.iloc[1]

#appending dataframes
#df1.append(df2)

#dropping a row by label
df = df.drop(0)

df
```

```python
#pandas series functionality

import pandas as pd
import numpy as np

s = pd.Series(np.random.randn(4))
print(s)

print(s.axes) # axes series
print(s.empty) # whether series is empty
print(s.ndim) # dimensions of series
print(s.size) # size of series
print(s.values) # values in series
print(s.head(2)) # head 2 values
print(s.tail(2)) # tail 2 values
```
